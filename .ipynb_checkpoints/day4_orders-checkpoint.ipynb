```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Baca data Customers
df = pd.read_csv('orders.csv')

# Convert 'order_date' and 'delivery_date' to datetime format
df['order_date'] = pd.to_datetime(df['order_date'])
df['delivery_date'] = pd.to_datetime(df['delivery_date'])

# Proses pembuatan kolom baru bernama delivery_time
delivery_time = df['delivery_date'] - df['order_date']

# Convert the timedelta to total seconds, then to days
delivery_time = delivery_time.apply(lambda x: x.total_seconds())

# Add delivery time in days to the dataframe (86400 seconds = 1 day)
df['delivery_time'] = round(delivery_time / 86400)

# Show the resulting dataframe
df.describe(include="all")

```
