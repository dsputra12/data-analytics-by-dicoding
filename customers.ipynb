```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Data yang belum dibersihkan
print("Data yang belum dibersihkan:\n")

#Customers
df = pd.read_csv('customers.csv')

#Menampikkan hasil tabel untuk 5 baris pertama
tabel_customer = df.head()

#Memberikan informasi terkait tabel customers secara keseluruhan
info_customer = df.info()
# print(tabel_customer)
print(info_customer)

# #Mencari data kosong
total_customer_kosong = df.isna().sum()
print(total_customer_kosong)

# #Mencari data yang duplikat
customer_duplikat = df.duplicated().sum()
print("Customer duplikat sebelum dibersihkan: ", customer_duplikat)

#Menampilkan deskripsi statistik (khusus data numerik)
deskripsi_customer = df.describe()
print(deskripsi_customer)

# #Menghapus data yang duplikat
customer_duplikat = df.drop_duplicates(inplace = True)
print("Customer yang duplikat setelah dibersihkan: ", customer_duplikat)

#Filter data untuk melihat data gender pada Customer yang kosong
gender_kosong = df[df.gender.isna()]
print(gender_kosong)

#Cek isi elemen pada kolom / variabel tertentu 
isi_gender = df.gender.value_counts()
print(isi_gender)

#Pembersihan data kosong (Cara 1)
df.fillna(value="Non Binary", inplace =True)
isi_gender = df.gender.value_counts()
print(isi_gender)

total_customer_kosong = df.isna().sum()
print(f"Jumlah gender setelah dibersihkan:\n{total_customer_kosong}")

#Pembersihan data kosong (Cara 2)
df['gender'] = df['gender'].fillna(value = "Non Binary Gaming")
isi_gender = df.gender.value_counts()
print(isi_gender)
total_customer_kosong = df.isna().sum()
print(f"Jumlah gender setelah dibersihkan:\n{total_customer_kosong}")

#Cek umur maksimum pada Customer
umur_max = df[df.age == df.age.max()]
print(umur_max)

#Mengganti elemen suatu kolom dengan replace()
df['age'] = df['age'].replace(df.age.max(), 70)
umur_max = df[df.age == df.age.max()]
deskripsi_customer = df.describe()
print(deskripsi_customer)
print(f"Umur setelah diperbaiki:\n{umur_max}")

#Setelah dicek, ternyata masih ada data yang belum akurat, yaitu data untuk customer id 261, maka kita perlu melakukan replace sekali lagi
df['age'] = df['age'].replace(df.age.max(), 50)
umur_max = df[df.age == df.age.max()]
deskripsi_customer = df.describe()
print(deskripsi_customer)
print(f"Umur setelah diperbaiki untuk ke-2 kalinya:\n{umur_max}")


```
