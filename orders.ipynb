```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Data yang belum dibersihkan
print("Data yang belum dibersihkan:\n")

#Orders
df2 = pd.read_csv('orders.csv')
tabel_orders = df2.head()
# print(tabel_orders)

#Memberikan informasi terkait tabel orders secara keseluruhan
info_orders = df2.info()
print(info_orders)

#Cek duplikasi data
orders_duplikat = df2.duplicated().sum()
print(orders_duplikat)

#Deskripsi statistik data Orders
deskripsi_orders = df2.describe()
print(deskripsi_orders)

#Mengubah tipe data dari suatu kolom menggunakan looping
target_columns = ['order_date', 'delivery_date']

for columns in target_columns:
    df2[columns] = pd.to_datetime(df2[columns])

info_order = df2.info()
print(f"Deskripsi Order setelah diperbaiki:\n{info_order}")

#Mengubah tipe data tanpa looping
df2['payment'] = pd.to_datetime(df2['payment'])
info_order = df2.info()
print(info_order)
```
